```python
import numpy as np
from matplotlib import pyplot as plt

#Create a Toy Data set

x=np.random.random((1000,1))
theta=np.array([2,3])
X=np.column_stack((x,np.ones(1000)))
Y=np.dot(X,theta)
#Y=Y.reshape(1000,1)

#Adding White Noise
Y+=0.1*np.random.randn(1000)

#Plotting the points
plt.figure(figsize=(20,10))
plt.scatter(x,Y)
```

```python
#Taking 70% of the data,Splitting X into X_train and X_test and similarly for Y.
X_train=X[0:700,:]
X_test=X[700:1000,:]
#This was where my error was i assumed Y as a matrix instead of a List
Y_train=Y[0:700]
Y_test=Y[700:1000]
#print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)

plt.figure(figsize=(20,10))
plt.scatter(X_train[:,0],Y_train)
```

# Gradient Descent:Training

```python
def loss_function(X,Y,tet):   
    Y_p=np.dot(X,tet)    
    J = np.sum(np.square(Y-Y_p))/len(Y)
    return J

def gradient_desc(theta,Y,X,itr,alpha):
    cost_hist=[0]*itr
    for i in range(itr):
        Y_p=np.dot(X,theta)
        loss=Y_p-Y 
        
        grad= np.sum(np.dot(loss,X))/len(Y)
        #grad1= np.sum(loss)/len(Y)
        theta=theta-alpha*grad
        #theta[1]=theta[1]-alpha*grad        
        J = loss_function(X,Y,theta)
        print(J)
       
        cost_hist[i]=J
        if(J>cost_hist[i-1]):
            alpha=alpha/10   
            
    return theta,cost_hist
theta_f,cost_hist=gradient_desc(theta,Y_train,X_train,1000,0.01)
print(theta_f)
```

```python
#Plotting the Training theta
plt.figure(figsize=(20,10))
plt.scatter(X_train[:,0],Y_train)

y_pred=np.dot(X_train,theta_f)
plt.plot(X_train[:,0],y_pred,'o',color="r")

#Finding Mean Squared Error
MSE=np.mean(np.square(Y_train-y_pred))
print(MSE)

```

# Gradient Descent: Testing

```python
plt.figure(figsize=(20,10))
plt.scatter(X_test[:,0],Y_test)

y_pred=np.dot(X_test,theta_f)
plt.plot(X_test[:,0],y_pred,'o',color="r")

#Finding Mean Squared Error
MSE=np.mean(np.square(Y_test-y_pred))
print(MSE)
```
