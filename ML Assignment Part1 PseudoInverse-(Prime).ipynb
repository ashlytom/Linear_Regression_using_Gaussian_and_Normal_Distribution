```python
import numpy as np
from matplotlib import pyplot as plt

#Create a Toy Data set

x=np.random.random((1000,1))
theta=np.array([2,3])
#Making the independant variable
X=np.column_stack((x,np.ones(1000)))
Y=np.dot(X,theta)
#Y=Y.reshape(1000,1)

#Adding White Noise to the dependant varibale
Y+=0.1*np.random.randn(1000)

#Plotting the points
plt.figure(figsize=(20,10))
plt.scatter(x,Y)
```

```python
#Taking 70% of the data,Splitting X into X_train and X_test and similarly for Y.
X_train=X[0:700,:]
X_test=X[700:1000,:]
Y_train=Y[0:700]
Y_test=Y[700:1000]
#print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)

plt.figure(figsize=(20,10))
plt.scatter(X_train[:,0],Y_train)
```

# Pseudo Inverse Training

```python
#Finding the pseudoInverse
xi=np.linalg.pinv(X_train)
theta = np.dot(xi,Y_train)

plt.figure(figsize=(20,10))
plt.scatter(X_train[:,0],Y_train)

#Plotting using the new theta
Y_pred=np.dot(X_train,theta)
plt.plot(X_train[:,0],Y_pred,'o',color='r')
m=theta[0]
c=theta[1]
print(m,c)

#Finding Mean Squared Error
MSE=np.mean(np.square(Y_train-Y_pred))
print(MSE)

```

# Pseudo Inverse:Testing

```python
#Plotting the testing Data
plt.figure(figsize=(20,10))
plt.scatter(X_test[:,0],Y_test)

#Plotting using the new theta
Y_pred=np.dot(X_test,theta)
plt.plot(X_test[:,0],Y_pred,'o',color='r')

#Finding Mean Squared Error
MSE=np.mean(np.square(Y_test-Y_pred))
print(MSE)
```
